```python
import matplotlib.pyplot as py

x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]
```

```python
py.title('Line Plot')
py.xlabel('x')
py.ylabel('y')
py.plot(x,y,'red',marker='o',dashes=[6,2],linewidth=2.5)
```

```python
z = [3,6,9,12,15,18]
py.title('Comparison')
py.xlabel('x')
py.ylabel('y')
py.plot(x,y,'red', x,z, 'green', dashes=[6,2], marker='o')
```

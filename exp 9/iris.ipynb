```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py

data = pd.read_csv("IRIS.csv")
print(data)
```

```python
s1 = data['sepal_length'][2:52]
p1 = data['petal_length'][2:52]
s2 = data['sepal_length'][52:102]
p2 = data['petal_length'][52:102]
s3 = data['sepal_length'][102:]
p3 = data['petal_length'][102:]

py.title('IRIS DATA SET')
py.xlabel('Sepal Length')
py.ylabel('Petal Length')
py.scatter(s1,p1,c='red')
py.scatter(s2,p2,c='green')
py.scatter(s3,p3,c='yellow')
py.legend(['Setosa','Versicolor','Virginica'],title='Types of flower')
py.show()
```
